```python
# Data save bat

import LocalEnvAL as le
import os

from LocalEnvAL.LEF import SLEF
def path_read(dir: str) -> list:
    files = os.listdir(dir)
    print(f"find {len(files)} files")
    struct_list = list(map(le.Structure.from_file, 
                           [ os.path.join(dir,file) for file in files 
                            if not os.path.isdir(os.path.join(dir, file))]))
   
    return struct_list

path = "./TestStructure/Si"
struct_list = path_read(path)
slef_list = list(map(le.SLEF, struct_list ))

for stct in slef_list:
    le.SLEF_save(stct)   
```

```python
# Data read bat 
import LocalEnvAL as le 
import os
path = "./output"
files_list = [os.path.join(path, file) for file in os.listdir(path)
              if file.endswith(".SLEF")]
slef_list = list(map(le.SLEF_read, files_list))
for slef in slef_list:
    print(slef)
```
