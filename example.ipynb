```python
import LocalEnvAL as le
import numpy as np
file_path = "TestStructure/BaYCuO/primitive_material_BaYCuO.cif"
struct = le.Structure.from_file(file_path)

print(struct)
# test parameter 
r_max = 6
dr: float = 0.01
sigma: float = 0.03

slef = le.SLEF(structure=struct, symmetrize=False)
print(slef)
print(slef.S)
```

```python
data = slef.results["O"][10]  # type: ignore
# warning: the "1" is a sub indice of Dict[Dict], 
# don't use number to get value like this, please use iterator
r = slef.r
fig_rdf = le.single_atom_rdf_plot(r, data=data, title= f"{slef.S.formula}: O")
fig = le.rcut_plot(r=r, rcut=slef.rcut, rdf_fig=fig_rdf, z_max=np.max(data))
fig.show()
```
